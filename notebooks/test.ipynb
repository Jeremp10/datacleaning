```python
import matplotlib.pyplot as plt

#Converting dates
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])

#Daily order volume
orders_cleaned = df_orders[df_orders['order_status'].isin(['delivered', 'invoiced', 'shipped'])].copy()
orders_cleaned['date'] = orders_cleaned['order_purchase_timestamp'].dt.date

daily_orders = orders_cleaned.groupby('date').size()

plt.figure(figsize=(14, 5))
daily_orders.plot()
plt.title('Daily Order Volume')
plt.ylabel('Number of Orders')
plt.xlabel('Date')
plt.show()

# Monthly trends
orders_cleaned['month'] = orders_cleaned['order_purchase_timestamp'].dt.to_period('M')
monthly_orders = orders_cleaned.groupby('month').size()
print("\nMonthly order volume:")
print(monthly_orders)
```

Filtering the data for a total of 12 months due to messy data prior to 2017

```python
#Filtering
start_date = '2017-07-01'
end_date = '2018-07-31'

orders_cleaned['order_purchase_timestamp'] = pd.to_datetime(orders_cleaned['order_purchase_timestamp'])

orders_filtered = orders_cleaned[(orders_cleaned['order_purchase_timestamp'] >= start_date)
                                  & (orders_cleaned['order_purchase_timestamp'] <= end_date)].copy()

print(f"Orders in analysis period: {len(orders_filtered)}")
```

```python
print(f"Date range: {orders_filtered['order_purchase_timestamp'].min()} to {orders_filtered['order_purchase_timestamp'].max()}")

# Check the data quality in this period
print("\nOrder status distribution:")
print(orders_filtered['order_status'].value_counts())
```

```python
orders_filtered['date'] = orders_filtered['order_purchase_timestamp'].dt.date
daily_orders_filtered = orders_filtered.groupby('date').size()

plt.Figure(figsize=(14,5))
daily_orders_filtered.plot()
plt.title('Daily Order Volume (Jul 2017 - Jul 2018)')
plt.ylabel('Number of Orders')
plt.xlabel('Date')
plt.axvline(pd.to_datetime('2017-11-24'), color='red', linestyle='--', alpha=0.5, label='Black Friday')
plt.axvline(pd.to_datetime('2017-12-25'), color='green', linestyle='--', alpha=0.5, label='Christmas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Merge with payments for this period
orders_filtered_payments = orders_filtered.merge(order_payments, on='order_id', how='left')

daily_revenue_filtered = orders_filtered_payments.groupby('date')['total_payment'].sum()

plt.figure(figsize=(14, 5))
daily_revenue_filtered.plot()
plt.title('Daily Revenue (Jul 2017 - Jul 2018)')
plt.ylabel('Revenue (BRL)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()

print(f"\nTotal revenue in period: BRL {daily_revenue_filtered.sum():,.2f}")
print(f"Average daily revenue: BRL {daily_revenue_filtered.mean():,.2f}")
print(f"Monthly average: BRL {daily_revenue_filtered.sum()/12:,.2f}")
```

```python
orders_filtered_ids = orders_filtered['order_id'].unique()
payments_filtered = df_payments[df_payments['order_id'].isin(orders_filtered_ids)].copy()

print("A breakdown of payment methods:")
print(payments_filtered['payment_type'].value_counts())
print("Percentage of payment method:")
print(payments_filtered['payment_type'].value_counts(normalize=True) * 100)

# Check installments (important for Brazil)
print("\nInstallment distribution:")
print(payments_filtered['payment_installments'].describe())
print("\nMost common installment counts:")
print(payments_filtered['payment_installments'].value_counts().head(10))
```

```python
payment_delays = {
    "credit_card": 3,
    "boleto": 1,
    "voucher": 2,
    "debit_card": 2,
    "not_defined": 3
}
```

```python
orders_with_payments = orders_filtered.merge(
    payments_filtered,
    on= 'order_id',
    how='left'
)

print(f"Total records after merge: {len(orders_with_payments)}")
print(f"Unique orders: {orders_with_payments['order_id'].nunique()}")

# Some orders have multiple payment methods - that's normal
orders_multi_payment = orders_with_payments.groupby('order_id').size()
print(f"\nOrders with multiple payments: {(orders_multi_payment > 1).sum()}")
```

```python
orders_with_payments['payment_delay_days']= orders_with_payments['payment_type'].map(payment_delays)


orders_with_payments['cash_receipt_date'] = pd.to_datetime(
    orders_with_payments['order_purchase_timestamp']
) + pd.to_timedelta(orders_with_payments['payment_delay_days'], unit='D')


orders_with_payments['cash_receipt_date'] = orders_with_payments['cash_receipt_date'].dt.date

print("\nSample of order vs cash receipt timing:")
print(orders_with_payments[['order_id', 'order_purchase_timestamp', 'payment_type',
                             'payment_value', 'payment_delay_days', 'cash_receipt_date']].head(10))
```

```python

daily_cash_inflows = orders_with_payments.groupby('cash_receipt_date')['payment_value'].sum().reset_index()
daily_cash_inflows.columns = ['date', 'cash_inflow']

# Sort by date
daily_cash_inflows = daily_cash_inflows.sort_values('date')

print("\nDaily cash inflows summary:")
print(daily_cash_inflows.describe())
print(f"\nTotal cash inflows: BRL {daily_cash_inflows['cash_inflow'].sum():,.2f}")

# Show first few days
print("\nFirst 10 days of cash inflows:")
print(daily_cash_inflows.head(10))
```

```python
orders_filtered_copy = orders_filtered.copy()
orders_filtered_copy['date'] = orders_filtered_copy['order_purchase_timestamp'].dt.date

order_revenue = orders_with_payments.groupby(
    pd.to_datetime(orders_with_payments['order_purchase_timestamp']).dt.date
)['payment_value'].sum().reset_index()
order_revenue.columns = ['date', 'order_revenue']

cash_revenue = daily_cash_inflows.copy()
cash_revenue.columns = ['date', 'cash_revenue']

revenue_comparison = order_revenue.merge(cash_revenue, on='date', how='outer').fillna(0)
revenue_comparison = revenue_comparison.sort_values('date')
```

```python

# Plot comparison
fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(revenue_comparison['date'], revenue_comparison['order_revenue'],
        label='Revenue (Order Date)', alpha=0.7, linewidth=1.5)
ax.plot(revenue_comparison['date'], revenue_comparison['cash_revenue'],
        label='Cash Inflow (Receipt Date)', alpha=0.7, linewidth=1.5)
ax.set_title('Revenue vs Cash Inflow: The Timing Gap', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Amount (BRL)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nKey insight: Notice the lag between when orders happen vs when cash arrives!")
```

```python
# Check for canceled orders in your period
canceled_orders = orders_filtered[orders_filtered['order_status'] == 'canceled']
print(f"\nCanceled orders: {len(canceled_orders)} ({len(canceled_orders)/len(orders_filtered)*100:.1f}%)")

# For canceled orders with payments, these become refunds (cash outflows)
canceled_with_payments = canceled_orders.merge(payments_filtered, on='order_id', how='inner')

if len(canceled_with_payments) > 0:
    # Assume refunds happen 7 days after order
    canceled_with_payments['refund_date'] = (
        pd.to_datetime(canceled_with_payments['order_purchase_timestamp']) +
        pd.timedelta(days=7)
    ).dt.date

    daily_refunds = canceled_with_payments.groupby('refund_date')['payment_value'].sum().reset_index()
    daily_refunds.columns = ['date', 'refund_outflow']

    print(f"Total refunds: BRL {daily_refunds['refund_outflow'].sum():,.2f}")
else:
    daily_refunds = pd.DataFrame(columns=['date', 'refund_outflow'])
    print("No canceled orders with payments found")
```

```python
# Combine inflows and refunds
cash_flow_revenue = daily_cash_inflows.merge(daily_refunds, on='date', how='outer').fillna(0)
cash_flow_revenue['net_cash_inflow'] = cash_flow_revenue['cash_inflow'] - cash_flow_revenue['refund_outflow']

# Ensure complete date range (fill missing dates with 0)
date_range = pd.date_range(start='2017-07-01', end='2018-07-31', freq='D')
complete_dates = pd.DataFrame({'date': date_range.date})

cash_flow_revenue = complete_dates.merge(cash_flow_revenue, on='date', how='left').fillna(0)

print("\nFinal daily cash inflow table:")
print(cash_flow_revenue.head(15))

print(f"\nTotal net cash inflows: BRL {cash_flow_revenue['net_cash_inflow'].sum():,.2f}")
print(f"Average daily cash inflow: BRL {cash_flow_revenue['net_cash_inflow'].mean():,.2f}")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
ax.fill_between(cash_flow_revenue['date'], 0, cash_flow_revenue['net_cash_inflow'],
                alpha=0.3, label='Net Cash Inflow')
ax.plot(cash_flow_revenue['date'], cash_flow_revenue['net_cash_inflow'],
        linewidth=1, alpha=0.8)
ax.set_title('Daily Net Cash Inflows (After Payment Processing Delays)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Cash Inflow (BRL)')
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
cash_flow_revenue.to_csv('cash_inflows_daily.csv', index=False)
print("\nSaved: cash_inflows_daily.csv")
```

```python
#assuming a 60% cogs
cogs_percentage = 0.6

orders_filtered_copy= orders_filtered.copy()
orders_filtered_copy['date'] = orders_filtered_copy['order_purchase_timestamp'].dt.date

#revenue per order date
order_revenue_for_cogs = orders_with_payments.groupby(
    pd.to_datetime(orders_with_payments['order_purchase_timestamp']).dt.date
)['payment_value'].sum().reset_index()
order_revenue_for_cogs.columns = ['date', 'revenue']

order_revenue_for_cogs['cogs'] = order_revenue_for_cogs['revenue'] * cogs_percentage


print("Daily COGS calculation:")
print(order_revenue_for_cogs.head(10))
print(f"\nTotal COGS for period: BRL {order_revenue_for_cogs['cogs'].sum():,.2f}")
print(f"Total Revenue for period: BRL {order_revenue_for_cogs['revenue'].sum():,.2f}")
print(f"Gross Profit: BRL {(order_revenue_for_cogs['revenue'].sum() - order_revenue_for_cogs['cogs'].sum()):,.2f}")
```

```python
#calculating monthly cogs
order_revenue_for_cogs['month']= pd.to_datetime(order_revenue_for_cogs['date']).dt.to_period('M')
monthly_cogs = order_revenue_for_cogs.groupby('month')['cogs'].sum().reset_index()
monthly_cogs['month'] = monthly_cogs['month'].dt.to_timestamp()

print(monthly_cogs)

inventory_purchases = []

for idx, row in monthly_cogs.iterrows():
    purchase_date = row['month'] - pd.Timedelta(days=15)

    purchase_amount = row['cogs']   * 1.5

    inventory_purchases.append({
        'purchase_date': purchase_date.date(),
        'inventory_purchase': purchase_amount
    })


inventory_df = pd.DataFrame(inventory_purchases)
print("\nInventory purchases based on COGS:")
print(inventory_df)
```

```python
#applying payment terms to suppliers (net 30)
inventory_df['payment_date'] = inventory_df['purchase_date'] + pd.Timedelta(days=30)
inventory_df['payment_date'] = pd.to_datetime(inventory_df['payment_date'])

inventory_cash_outflows = inventory_df[['payment_date', 'inventory_purchase']].copy()
inventory_cash_outflows.columns = ['date', 'inventory_payment']

print(inventory_cash_outflows)
```

```python
#adding shipping and freight costs

order_items_filtered = df_order_items[df_order_items['order_id'].isin(orders_filtered_ids)].copy()

shipping_per_order = order_items_filtered.groupby('order_id')['freight_value'].sum().reset_index()

#merge with order dates
orders_shipping = orders_filtered[['order_id', 'order_purchase_timestamp']].merge(
    shipping_per_order, on='order_id', how='left')

orders_shipping['date'] = orders_shipping['order_purchase_timestamp'].dt.date

daily_shipping = orders_shipping.groupby('date')['freight_value'].sum().reset_index()
daily_shipping.columns = ['date', 'shipping_cost']

print(daily_shipping.describe())
```

```python
#adding payment processing fees (assume 3-4%)
payment_processing_fee_rate = 0.035

cash_flow_revenue_copy = cash_flow_revenue.copy()
cash_flow_revenue_copy['processing_fee'] = cash_flow_revenue_copy['cash_inflow'] * payment_processing_fee_rate

print(f"Total processing fees: BRL {cash_flow_revenue_copy['processing_fee'].sum():,.2f}")
print(f"Average daily fee: BRL {cash_flow_revenue_copy['processing_fee'].mean():,.2f}")
```

```python
#ad and marketing expenses (assume 20%)
marketing_percentage = 0.20

marketing_daily = order_revenue_for_cogs[['date', 'revenue']].copy()
marketing_daily['marketing_expense'] = marketing_daily['revenue'] * marketing_percentage

print(marketing_daily.head(10))
```

```python
#adding platform/marketplace fees (assume 10%)
platform_fee_percentage = 0.10

platform_fees = order_revenue_for_cogs[['date', 'revenue']].copy()
platform_fees['platform_fee'] = platform_fees['revenue'] * platform_fee_percentage

print("\nPlatform fees:")
print(f"Total platform fees: BRL {platform_fees['platform_fee'].sum():,.2f}")
```

```python
#adding fixed operating costs
monthly_fixed_costs = 50000

# Create daily fixed costs (divide monthly by ~30)
date_range = pd.date_range(start='2017-07-01', end='2018-07-31', freq='D')
fixed_costs_df = pd.DataFrame({
    'date': date_range.date,
    'fixed_costs': monthly_fixed_costs / 30  # Daily allocation
})

print("\nFixed costs:")
print(f"Monthly fixed costs: BRL {monthly_fixed_costs:,.2f}")
print(f"Daily fixed costs: BRL {monthly_fixed_costs/30:,.2f}")
print(f"Total for period (13 months): BRL {len(fixed_costs_df) * (monthly_fixed_costs/30):,.2f}")
```

```python
# Merge all cost components with correct column names
cash_outflows = complete_dates.copy()

cash_outflows = cash_outflows.merge(inventory_cash_outflows[['date', 'inventory_payment']], on='date', how='left')
cash_outflows = cash_outflows.merge(daily_shipping[['date', 'shipping_cost']], on='date', how='left')
cash_outflows = cash_outflows.merge(cash_flow_revenue_copy[['date', 'processing_fee']], on='date', how='left')
cash_outflows = cash_outflows.merge(marketing_daily[['date', 'marketing_expense']], on='date', how='left')  # Fixed
cash_outflows = cash_outflows.merge(platform_fees[['date', 'platform_fee']], on='date', how='left')
cash_outflows = cash_outflows.merge(fixed_costs_df[['date', 'fixed_costs']], on='date', how='left')

# Fill NaN with 0
cash_outflows = cash_outflows.fillna(0)

# Calculate total daily outflows with correct column name
cash_outflows['total_outflows'] = (
    cash_outflows['inventory_payment'] +
    cash_outflows['shipping_cost'] +
    cash_outflows['processing_fee'] +
    cash_outflows['marketing_expense'] +  # Fixed
    cash_outflows['platform_fee'] +
    cash_outflows['fixed_costs']
)


print(f"\nShape: {cash_outflows.shape}")
print("\nFirst few rows:")
print(cash_outflows.head())

print("\nTotal outflows by category:")
print(f"Inventory payments:  BRL {cash_outflows['inventory_payment'].sum():,.2f}")
print(f"Shipping:            BRL {cash_outflows['shipping_cost'].sum():,.2f}")
print(f"Processing fees:     BRL {cash_outflows['processing_fee'].sum():,.2f}")
print(f"Marketing:           BRL {cash_outflows['marketing_expense'].sum():,.2f}")
print(f"Platform fees:       BRL {cash_outflows['platform_fee'].sum():,.2f}")
print(f"Fixed costs:         BRL {cash_outflows['fixed_costs'].sum():,.2f}")
print(f"")
print(f"TOTAL OUTFLOWS:      BRL {cash_outflows['total_outflows'].sum():,.2f}")
```

```python
cost_summary ={
    'Inventory payments': cash_outflows['inventory_payment'].sum(),
    'Shipping': cash_outflows['shipping_cost'].sum(),
    'Processing fees': cash_outflows['processing_fee'].sum(),
    'Marketing': cash_outflows['marketing_expense'].sum(),
    'Platform fees': cash_outflows['platform_fee'].sum(),
    'Fixed costs': cash_outflows['fixed_costs'].sum()
}
```

```python
plt.figure(figsize=(10, 7))
plt.pie(cost_summary.values(), labels=cost_summary.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Cash Outflow Breakdown', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
