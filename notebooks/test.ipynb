##### Loading the first tables order_items, orders, and payments



```python
import pandas as pd
df_orders = pd.read_csv("../data_raw/olist_orders_dataset.csv")
df_orders.head()
```

```python
df_order_items = pd.read_csv("../data_raw/olist_order_items_dataset.csv")
df_order_items.head()
```

```python
df_payments = pd.read_csv("../data_raw/olist_order_payments_dataset.csv")
df_payments.head()
```

```python
print("Missing values in orders dataset:")
print(df_orders.isnull().sum())

print("\nMissing values in order items dataset:")
print(df_order_items.isnull().sum())

print("\nMissing values in payments dataset:")
print(df_payments.isnull().sum())
```

```python
#Checking order statuses
print("Order statuses:")
print(df_orders['order_status'].value_counts())
```

```python

delivered_orders = df_orders[df_orders['order_status'].isin(['delivered', 'invoiced', 'shipped'])]
print(f"\nUsable orders for cash flow: {len(delivered_orders)} out of {len(df_orders)}")
```

```python
#Understanding relationships
payment_orders = df_payments['order_id'].unique()
orders = df_orders['order_id'].unique()
print(f"Payment without order: {len(set(payment_orders) - set(orders))}")
```

```python
#Checking payment installments
print("Payment installments:")
print(df_payments['payment_installments'].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt

#Converting dates
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])

#Daily order volume
orders_cleaned = df_orders[df_orders['order_status'].isin(['delivered', 'invoiced', 'shipped'])].copy()
orders_cleaned['date'] = orders_cleaned['order_purchase_timestamp'].dt.date

daily_orders = orders_cleaned.groupby('date').size()

plt.figure(figsize=(14, 5))
daily_orders.plot()
plt.title('Daily Order Volume')
plt.ylabel('Number of Orders')
plt.xlabel('Date')
plt.show()

# Monthly trends
orders_cleaned['month'] = orders_cleaned['order_purchase_timestamp'].dt.to_period('M')
monthly_orders = orders_cleaned.groupby('month').size()
print("\nMonthly order volume:")
print(monthly_orders)
```

```python
#checking payment types
print("\nPayment types:")
print(df_payments['payment_type'].value_counts())
```

```python
# Payment values
print("\nPayment value statistics:")
print(df_payments['payment_value'].describe())

# Join to see total payment per order
order_payments = df_payments.groupby('order_id')['payment_value'].sum().reset_index()
order_payments.columns = ['order_id', 'total_payment']

print("\nAverage order value:")
print(order_payments['total_payment'].describe())
```

```python
#merge payments with orders
orders_with_payments = pd.merge(orders_cleaned, order_payments, on='order_id', how='left')
daily_revenue = orders_with_payments.groupby('date')['total_payment'].sum()


plt.figure(figsize=(14, 5))
daily_revenue.plot()
plt.title('Daily Revenue')
plt.ylabel('Revenue (BRL)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()

print("\nRevenue statistics:")
print(f"Total revenue: BRL {daily_revenue.sum():,.2f}")
print(f"Average daily revenue: BRL {daily_revenue.mean():,.2f}")
print(f"Date range: {daily_revenue.index.min()} to {daily_revenue.index.max()}")
```
