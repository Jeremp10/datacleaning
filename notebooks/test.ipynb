```python
payment_delays = {
    "credit_card": 3,
    "boleto": 1,
    "voucher": 2,
    "debit_card": 2,
    "not_defined": 3
}
```

```python
orders_with_payments = orders_filtered.merge(
    payments_filtered,
    on= 'order_id',
    how='left'
)

print(f"Total records after merge: {len(orders_with_payments)}")
print(f"Unique orders: {orders_with_payments['order_id'].nunique()}")

# Some orders have multiple payment methods - that's normal
orders_multi_payment = orders_with_payments.groupby('order_id').size()
print(f"\nOrders with multiple payments: {(orders_multi_payment > 1).sum()}")
```

```python
orders_with_payments['payment_delay_days']= orders_with_payments['payment_type'].map(payment_delays)


orders_with_payments['cash_receipt_date'] = pd.to_datetime(
    orders_with_payments['order_purchase_timestamp']
) + pd.to_timedelta(orders_with_payments['payment_delay_days'], unit='D')


orders_with_payments['cash_receipt_date'] = orders_with_payments['cash_receipt_date'].dt.date

print("\nSample of order vs cash receipt timing:")
print(orders_with_payments[['order_id', 'order_purchase_timestamp', 'payment_type',
                             'payment_value', 'payment_delay_days', 'cash_receipt_date']].head(10))
```

```python

daily_cash_inflows = orders_with_payments.groupby('cash_receipt_date')['payment_value'].sum().reset_index()
daily_cash_inflows.columns = ['date', 'cash_inflow']

# Sort by date
daily_cash_inflows = daily_cash_inflows.sort_values('date')

print("\nDaily cash inflows summary:")
print(daily_cash_inflows.describe())
print(f"\nTotal cash inflows: BRL {daily_cash_inflows['cash_inflow'].sum():,.2f}")

# Show first few days
print("\nFirst 10 days of cash inflows:")
print(daily_cash_inflows.head(10))
```

```python
orders_filtered_copy = orders_filtered.copy()
orders_filtered_copy['date'] = orders_filtered_copy['order_purchase_timestamp'].dt.date

order_revenue = orders_with_payments.groupby(
    pd.to_datetime(orders_with_payments['order_purchase_timestamp']).dt.date
)['payment_value'].sum().reset_index()
order_revenue.columns = ['date', 'order_revenue']

cash_revenue = daily_cash_inflows.copy()
cash_revenue.columns = ['date', 'cash_revenue']

revenue_comparison = order_revenue.merge(cash_revenue, on='date', how='outer').fillna(0)
revenue_comparison = revenue_comparison.sort_values('date')
```

```python

# Plot comparison
fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(revenue_comparison['date'], revenue_comparison['order_revenue'],
        label='Revenue (Order Date)', alpha=0.7, linewidth=1.5)
ax.plot(revenue_comparison['date'], revenue_comparison['cash_revenue'],
        label='Cash Inflow (Receipt Date)', alpha=0.7, linewidth=1.5)
ax.set_title('Revenue vs Cash Inflow: The Timing Gap', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Amount (BRL)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nKey insight: Notice the lag between when orders happen vs when cash arrives!")
```

```python
# Check for canceled orders in your period
canceled_orders = orders_filtered[orders_filtered['order_status'] == 'canceled']
print(f"\nCanceled orders: {len(canceled_orders)} ({len(canceled_orders)/len(orders_filtered)*100:.1f}%)")

# For canceled orders with payments, these become refunds (cash outflows)
canceled_with_payments = canceled_orders.merge(payments_filtered, on='order_id', how='inner')

if len(canceled_with_payments) > 0:
    # Assume refunds happen 7 days after order
    canceled_with_payments['refund_date'] = (
        pd.to_datetime(canceled_with_payments['order_purchase_timestamp']) +
        pd.timedelta(days=7)
    ).dt.date

    daily_refunds = canceled_with_payments.groupby('refund_date')['payment_value'].sum().reset_index()
    daily_refunds.columns = ['date', 'refund_outflow']

    print(f"Total refunds: BRL {daily_refunds['refund_outflow'].sum():,.2f}")
else:
    daily_refunds = pd.DataFrame(columns=['date', 'refund_outflow'])
    print("No canceled orders with payments found")
```

```python
# Combine inflows and refunds
cash_flow_revenue = daily_cash_inflows.merge(daily_refunds, on='date', how='outer').fillna(0)
cash_flow_revenue['net_cash_inflow'] = cash_flow_revenue['cash_inflow'] - cash_flow_revenue['refund_outflow']

# Ensure complete date range (fill missing dates with 0)
date_range = pd.date_range(start='2017-07-01', end='2018-07-31', freq='D')
complete_dates = pd.DataFrame({'date': date_range.date})

cash_flow_revenue = complete_dates.merge(cash_flow_revenue, on='date', how='left').fillna(0)

print("\nFinal daily cash inflow table:")
print(cash_flow_revenue.head(15))

print(f"\nTotal net cash inflows: BRL {cash_flow_revenue['net_cash_inflow'].sum():,.2f}")
print(f"Average daily cash inflow: BRL {cash_flow_revenue['net_cash_inflow'].mean():,.2f}")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
ax.fill_between(cash_flow_revenue['date'], 0, cash_flow_revenue['net_cash_inflow'],
                alpha=0.3, label='Net Cash Inflow')
ax.plot(cash_flow_revenue['date'], cash_flow_revenue['net_cash_inflow'],
        linewidth=1, alpha=0.8)
ax.set_title('Daily Net Cash Inflows (After Payment Processing Delays)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Cash Inflow (BRL)')
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

### Phase 4
#### Building cash flow model

```python
# Standardize date types before merging
cash_flow_revenue['date'] = pd.to_datetime(cash_flow_revenue['date'])
cash_outflows['date'] = pd.to_datetime(cash_outflows['date'])

# Verify they match
print("Before merge - checking dtypes:")
print(f"cash_flow_revenue['date']: {cash_flow_revenue['date'].dtype}")
print(f"cash_outflows['date']: {cash_outflows['date'].dtype}")

# Now merge - should work
daily_cash_flow = cash_flow_revenue[['date', 'net_cash_inflow']].merge(
    cash_outflows[['date', 'inventory_payment', 'shipping_cost',
                   'processing_fee', 'marketing_expense',
                   'platform_fee', 'fixed_costs', 'total_outflows']],
    on='date',
    how='outer'
).fillna(0)

# Sort by date
daily_cash_flow = daily_cash_flow.sort_values('date').reset_index(drop=True)

# Calculate net daily cash flow
daily_cash_flow['net_daily_cash_flow'] = (
    daily_cash_flow['net_cash_inflow'] -
    daily_cash_flow['total_outflows']
)

print("\n Merge successful!")
print(f"Shape: {daily_cash_flow.shape}")
print("\nDaily cash flow table:")
print(daily_cash_flow.head(10))
```

```python
# For our model, assume BRL 500,000 opening balance
OPENING_BALANCE = 500000

# Calculate cumulative cash balance
daily_cash_flow['running_balance'] = (
    OPENING_BALANCE +
    daily_cash_flow['net_daily_cash_flow'].cumsum()
)

print("Running cash balance:")
print(daily_cash_flow[['date', 'net_cash_inflow', 'total_outflows',
                        'net_daily_cash_flow', 'running_balance']].head(20))

# Key statistics
print(f"\nOpening balance:      BRL {OPENING_BALANCE:,.2f}")
print(f"Closing balance:      BRL {daily_cash_flow['running_balance'].iloc[-1]:,.2f}")
print(f"Lowest cash point:    BRL {daily_cash_flow['running_balance'].min():,.2f}")
print(f"Highest cash point:   BRL {daily_cash_flow['running_balance'].max():,.2f}")
print(f"Date of lowest cash:  {daily_cash_flow.loc[daily_cash_flow['running_balance'].idxmin(), 'date']}")
print(f"Date of highest cash: {daily_cash_flow.loc[daily_cash_flow['running_balance'].idxmax(), 'date']}")
```

#### Identifying cash crunches

```python
#identify cash crunches that fall below the danger zone
danger_threshold = 100000
critical_threshold = 50000

daily_cash_flow['cash_status'] = 'healthy'
daily_cash_flow.loc[daily_cash_flow['running_balance'] < danger_threshold, 'cash_status'] = 'warning'
daily_cash_flow.loc[daily_cash_flow['running_balance'] < critical_threshold, 'cash_status'] = 'critical'
daily_cash_flow.loc[daily_cash_flow['running_balance'] < 0, 'cash_status'] = 'insolvent'

print("Cash status distribution:")
print(daily_cash_flow['cash_status'].value_counts())


#show warning periods
warning_periods = daily_cash_flow[daily_cash_flow['cash_status'].isin(['warning', 'critical', 'insolvent'])]
if len(warning_periods) > 0:
    print(f"\n  Days below danger threshold: {len(warning_periods)}")
    print(warning_periods[['date', 'running_balance', 'cash_status']].head(10))
else:
    print("\n No cash crunches detected with current opening balance!")
```

```python
# Aggregate to monthly for cleaner analysis
daily_cash_flow['month'] = pd.to_datetime(daily_cash_flow['date']).dt.to_period('M')

monthly_summary = daily_cash_flow.groupby('month').agg(
    total_inflows=('net_cash_inflow', 'sum'),
    total_outflows=('total_outflows', 'sum'),
    net_cash_flow=('net_daily_cash_flow', 'sum'),
    closing_balance=('running_balance', 'last'),
    lowest_balance=('running_balance', 'min')
).reset_index()

monthly_summary['month'] = monthly_summary['month'].dt.to_timestamp()

print("\nMonthly Cash Flow Summary:")
print(monthly_summary.to_string(index=False))
```

```python
fig, axes = plt.subplots(3, 1, figsize=(16, 18))

# Chart 1: Daily Cash Inflows vs Outflows
ax1 = axes[0]
ax1.fill_between(daily_cash_flow['date'], 0, daily_cash_flow['net_cash_inflow'],
                 alpha=0.4, color='green', label='Cash Inflows')
ax1.fill_between(daily_cash_flow['date'], 0, -daily_cash_flow['total_outflows'],
                 alpha=0.4, color='red', label='Cash Outflows')
ax1.axhline(y=0, color='black', linewidth=0.8, linestyle='-')
ax1.set_title('Daily Cash Inflows vs Outflows', fontsize=14, fontweight='bold')
ax1.set_ylabel('Amount (BRL)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Chart 2: Running Cash Balance
ax2 = axes[1]
# Color the line based on health status
ax2.plot(daily_cash_flow['date'], daily_cash_flow['running_balance'],
         color='blue', linewidth=1.5, label='Cash Balance')
ax2.axhline(y=danger_threshold, color='orange', linewidth=1.5,
            linestyle='--', label=f'Warning (BRL {danger_threshold:,})')
ax2.axhline(y=critical_threshold, color='red', linewidth=1.5,
            linestyle='--', label=f'Critical (BRL {critical_threshold:,})')
ax2.axhline(y=0, color='black', linewidth=1, linestyle='-')
ax2.fill_between(daily_cash_flow['date'], 0, daily_cash_flow['running_balance'],
                 where=daily_cash_flow['running_balance'] >= danger_threshold,
                 alpha=0.2, color='green', label='Healthy zone')
ax2.fill_between(daily_cash_flow['date'], 0, daily_cash_flow['running_balance'],
                 where=daily_cash_flow['running_balance'] < danger_threshold,
                 alpha=0.2, color='red', label='Danger zone')
ax2.set_title('Running Cash Balance', fontsize=14, fontweight='bold')
ax2.set_ylabel('Cash Balance (BRL)')
ax2.legend(loc='upper left')
ax2.grid(True, alpha=0.3)

# Chart 3: Monthly cash flow

ax3 = axes[2]
colors = ['green' if x >= 0 else 'red' for x in monthly_summary['net_cash_flow']]
bars = ax3.bar(monthly_summary['month'], monthly_summary['net_cash_flow'],
               color=colors, alpha=0.7, width=20)
ax3.axhline(y=0, color='black', linewidth=0.8)
ax3.set_title('Monthly Net Cash Flow', fontsize=14, fontweight='bold')
ax3.set_ylabel('Net Cash Flow (BRL)')
ax3.set_xlabel('Month')


# Add value labels on bars
for bar, val in zip(bars, monthly_summary['net_cash_flow']):
    ax3.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
             f'BRL {val:,.0f}', ha='center',
             va='bottom' if val >= 0 else 'top', fontsize=8)

ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('cash_flow_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n Dashboard saved as cash_flow_dashboard.png")
```

```python
#What if analysis
scenarios = {
    'Base Case': 1.0,          # Current trajectory
    'Sales Drop 20%': 0.8,     # Revenue falls 20%
    'Sales Growth 20%': 1.2,   # Revenue grows 20%
    'Sales Drop 40%': 0.6      # Severe downturn
}

fig, ax = plt.subplots(figsize=(16, 7))

colors = ['blue', 'red', 'green', 'orange']

for (scenario_name, multiplier), color in zip(scenarios.items(), colors):
    # Adjust inflows by scenario multiplier
    scenario_inflows = daily_cash_flow['net_cash_inflow'] * multiplier
    scenario_net = scenario_inflows - daily_cash_flow['total_outflows']
    scenario_balance = OPENING_BALANCE + scenario_net.cumsum()

    ax.plot(daily_cash_flow['date'], scenario_balance,
            label=scenario_name, color=color,
            linewidth=2, alpha=0.8)

ax.axhline(y=danger_threshold, color='black', linewidth=1,
           linestyle='--', alpha=0.5, label='Warning threshold')
ax.axhline(y=0, color='black', linewidth=1)
ax.fill_between(daily_cash_flow['date'], 0, danger_threshold,
                alpha=0.05, color='red')
ax.set_title('Cash Flow Scenario Analysis', fontsize=14, fontweight='bold')
ax.set_ylabel('Cash Balance (BRL)')
ax.set_xlabel('Date')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('scenario_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```
